```python
import yfinance as yf
import plotly.express as px
```

```python
aapl = yf.Ticker('AAPL')
df = aapl.history(period = '10y')
```

```python
missing = df.isnull().sum()
print(missing)

nan = df.isna().sum()
print(nan)

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df < lower_bound) | (df > upper_bound)]
print(outliers)
```

```python
fig = px.scatter(df, y = "Close", title = 'Closing Price of AAPL Over the Past 10 Years')
# fig.show()
```

```python
csvFile1 = '../Data/AAPL_Close_Price.csv'
df['Close'].to_csv(csvFile1, index = True)

csvFile2 = '../Data/AAPL_Stock_Info.csv'
df.to_csv(csvFile2, index = True)
```
